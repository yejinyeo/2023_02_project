```python
import pandas as pd

# CSV 파일 경로 지정
csv_file_path = '/content/drive/MyDrive/project/project_cs1/coupang_data/sentimental_output.csv'  # 실제 파일 경로로 변경해야 합니다.

# CSV 파일을 데이터프레임으로 불러오기
df = pd.read_csv(csv_file_path)
df['percentage']=0
df['rescore']=0


df
```

```python
import re
index_negative = []
for i in range(50):
    if df['evaluation'][i] .find('부정') == 1:
        text = df['evaluation'][i]
        matches = re.findall(r'\d+\.\d+', text)
        num = float(matches[0])
        df['percentage'][i] = 1-num
```

```python
import re
index_positive = []
for i in range(50):
    if df['evaluation'][i] .find('긍정') == 1:
        text = df['evaluation'][i]
        matches = re.findall(r'\d+\.\d+', text)
        num = float(matches[0])
        df['percentage'][i] = num
```

```python
df
```

```python
for i in range(50):
    num = float(df['percentage'][i])
    if num < 0.40:
        df['rescore'][i] = 1
    elif num < 0.50:
        df['rescore'][i] = 2
    elif num < 0.60:
        df['rescore'][i] = 3
    elif num < 0.80:
        df['rescore'][i] = 4
    else:
        df['rescore'][i] = 5

df
```
